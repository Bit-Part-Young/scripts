### 热图、关联图绘制

#### matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(0)
data = np.random.randn(30, 5)
columns = ["Feature1", "Feature2", "Feature3", "Feature4", "Feature5"]
df = pd.DataFrame(data, columns=columns)

corr = df.corr()  # 计算 Pearson 相关系数矩阵
```

```python
plt.matshow(corr)
```

#### seaborn

```python
import seaborn as sns

sns.heatmap(data=corr, 
            annot=True,   # 在单元格中显示相关系数数值
            cmap='coolwarm',
            # vmin=-1, vmax=1,  # 颜色映射范围
            square=True,  # 方形单元格
            linewidths=.5,  # 单元格之间的线宽
            # cbar_kws={"shrink": 1.0}  # colorbar 参数
           )


plt.show()
```

#### biokit

饼状关联图绘制：

```python
from biokit.viz import corrplot

c = corrplot.Corrplot(df)
c.params['colorbar.shrink'] = 1.0
```

```python
# 特征标签不是按顺序排列的
c.plot(method="pie", shrink=0.9, grid=True, upper="pie", colorbar=True, )
```

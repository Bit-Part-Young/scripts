### 使用 ase 计算 rdf

```python
from ase.build import bulk
from ase.atoms import Atoms
from ase.geometry.analysis import Analysis


def rdf_cal(atoms: Atoms, scale: int, rmax: float):

    atoms = atoms * scale

    ana = Analysis(atoms)

    result = ana.get_rdf(12, 100, return_dists=True)

    return result
```

```python
atoms = bulk("Ni", "fcc", a=3.52, cubic=True)
result = rdf_cal(
    atoms=atoms,
    scale=8,
    rmax=12,
)
result
```

```python
import matplotlib.pyplot as plt

# 绘制 RDF
fig, ax = plt.subplots()

ax.plot(result[0][1], result[0][0], label="Ni")
ax.set_xlabel("Distance")
ax.set_ylabel("RDF")

ax.legend()
```

```python
atoms = bulk("Fe", "bcc", a=2.87, cubic=True)

result = rdf_cal(
    atoms=atoms,
    scale=10,
    rmax=12,
)
result
```

### pymatviz 程序使用

元素周期表绘制

```python
# https://gist.github.com/lan496/ebfb9722e0c7af575de0958696ffda73
import numpy as np
import ase.data
from pymatgen.core.periodic_table import ElementBase
import pandas as pd
from pymatviz.ptable import ptable_heatmap
from pymatviz.utils import df_ptable

series = pd.Series(
    {
        ElementBase.from_Z(Z).symbol: val
        for Z, val in enumerate(ase.data.vdw_radii)
        if Z >= 1
    }
)
series.rename_axis("symbol", inplace=True)
series = series.reindex(df_ptable.index, fill_value=np.nan)

ax = ptable_heatmap(series, fmt=".2f", cbar_title="VdW radius")
```

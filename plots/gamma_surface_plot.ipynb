### GSFE gamma surface 绘制

数据来源：https://www.ctcms.nist.gov/potentials/atomman/tutorial/4.5._Gamma_surface_plotting.html

```python
import numpy as np
import pandas as pd

csv_fn = "gamma_surface.csv"
df = pd.read_csv(csv_fn)

df.shape
```

```python
df.head()
```

```python
x = df["x"]
y = df["y"]
gamma = df["gamma"]
```

```python
x = np.arange(0, 0.999, 1/30)
y = np.arange(0, 0.999, 1/30)

X, Y = np.meshgrid(x, y)
print(X.shape, Y.shape)
print(X[0])
print(Y[0])
```

```python
gamma_array = np.array(gamma).reshape(30, 30)
```

```python
import matplotlib.pyplot as plt

plt.contourf(X, Y, gamma_array, 20)
```

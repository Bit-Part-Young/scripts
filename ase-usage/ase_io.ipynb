### ase.io 模块

```python
%%bash

cat Nb5Si3_alpha.vasp
```

```python
from ase.io import read, write

structure_fn = "Nb5Si3_alpha.vasp"
structure = read(
    filename=structure_fn,
    format="vasp",
)
structure
```

```python
output_vasp_fn = "POSCAR"
write(
    output_vasp_fn,
    images=structure,
    format="vasp",
    direct=True,
    sort=True,
)
```

```python
%%bash

cat POSCAR
```

```python
output_xyz_fn = "Nb5Si3_alpha.xyz"
write(
    output_xyz_fn,
    images=structure,
    format="xyz",
)
```

```python
%%bash

cat Nb5Si3_alpha.xyz
```

```python
output_extxyz_fn = "Nb5Si3_alpha_ext.xyz"
# structure.info = {"config_type": "Nb5Si3_alpha"}
write(
    output_extxyz_fn,
    images=structure,
    format="extxyz",
)
```

```python
%%bash

cat Nb5Si3_alpha_ext.xyz
```

```python
output_lammps_data_fn = "Nb5Si3_alpha.lammps-data"
ele_list = ["Si", "Nb"]
write(
    output_lammps_data_fn,
    images=structure,
    format="lammps-data",
    specorder=ele_list,  # 指定 atom type 顺序；默认按照元素符号字母排序
    units="metal",
    atom_style="atomic",
)
```

```python
%%bash

cat Nb5Si3_alpha.lammps-data
```

#### VASP OUTCAR 转换 extxyz 格式

可获取 OUTCAR 中的所有离子步构型的原子位置、能量、受力、应力等信息

>reference: https://gist.github.com/simonbatzner/c2b05d38789b67f6fe5d3c75a4f2223d

```python
from ase.io import read, write

vasp_out_fn = "./OUTCAR"

images = read(
    filename=vasp_out_fn,
    format="vasp-out",
    index=":",
)
len(images)
```

```python
images[0]
```

```python
images[0].get_positions()
```

```python
images[0].get_potential_energy()
```

```python
images[0].get_forces()
```

```python
images[0].get_stress()
```

```python
extxyz_fn = "outcar.xyz"
write(filename=extxyz_fn, images=images, format="extxyz", append=True)
```

```python
from ase.eos import EquationOfState
from ase.units import kJ
import pandas as pd

csv_fn = "eng_vol.dat"
df = pd.read_csv(csv_fn, sep=None, engine="python")

df_volume = df.iloc[:, 1]
df_energy = df.iloc[:, 2]
```

```python
eos = EquationOfState(df_volume, df_energy, eos="birchmurnaghan")
v0, e0, B = eos.fit()
print(f"v0 = {v0:.3f}")
print(f"e0 = {e0:.3f}")
print(f"B = {B / kJ * 1.0e24:.1f} GPa")
```

```python
ax = eos.plot()
ax.set_title(label=None)
```

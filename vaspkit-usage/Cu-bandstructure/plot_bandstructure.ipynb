```python
import numpy as np

band_file = "REFORMATTED_BAND.dat"

band_data = []
with open(band_file, "r") as f:
    lines = f.readlines()

for line in lines[1:]:
    band_data.append([float(x) for x in line.strip().split()])

band_array = np.array(band_data)
band_array
```

```python
klable_fn = "KLABELS"

klable_list = []
with open(klable_fn, "r") as f:
    lines = f.readlines()

for line in lines[1:]:
    line = line.strip().split()
    if line:
        klable_list.append(line)
    else:
        break

klable_list
```

```python
import matplotlib.pyplot as plt
from spt.plot_params import set_roman_plot_params
from matplotlib.axes import Axes
from matplotlib.ticker import MultipleLocator

set_roman_plot_params()
fig, ax = plt.subplots()
ax: Axes

for i in range(band_array.shape[1] - 1):
    ax.plot(band_array[:, 0], band_array[:, i + 1])

ax.axhline(y=0, color="black", linewidth=0.5)

x_list = []
for j in range(len(klable_list)):
    x = float(klable_list[j][1])
    x_list.append(x)
    ax.axvline(x=x, color="black", linestyle="--", linewidth=0.5)

ax.set_xlim(x_list[0], x_list[-1])
ax.set_ylim(-10, 10)

xlabels = [
    klable_list[k][0] if klable_list[k][0] != "Gamma" else "$\Gamma$"
    for k in range(len(klable_list))
]

ax.set_xticks(x_list, labels=xlabels)

ax.set_ylabel("Energy (eV)")
```

EOS状态方程拟合

```python
from pymatgen.analysis.eos import BirchMurnaghan
import pandas as pd


# fcc Cu 的原胞体积和能量数据
# 体模量 B 实验值为 140 GPa
data = pd.read_csv("eng_vol.dat", sep=" ")
eng_data = data["energy(eV/cell)"]
vol_data = data["vol/atom"]

eos = BirchMurnaghan(volumes=vol_data, energies=eng_data)
eos.fit()
```

```python
eos.e0
```

```python
eos.b0_GPa
```

```python
eos.v0
```

```python
eos.results
```

```python
eos.plot(width=10, height=8, text=None)
# eos.plot(text=None)
```

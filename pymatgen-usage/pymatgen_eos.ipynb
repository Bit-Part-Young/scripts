### EOS 状态方程拟合

```python
from pymatgen.analysis.eos import BirchMurnaghan
import pandas as pd

# FCC Cu DFT 计算原胞体积和能量数据
# 体模量 B 实验值为 140 GPa
csv_fn = "eng_vol.dat"
df = pd.read_csv(csv_fn, sep=None, engine="python")
df_volume = df.iloc[:, 1]
df_energy = df.iloc[:, 2]

eos = BirchMurnaghan(volumes=df_volume, energies=df_energy)
eos.fit()
```

### pymatgen Vasprun 类

```python
from pymatgen.io.vasp.outputs import Vasprun

vasprun = Vasprun("vasprun.xml")
```

`ionic_step_skip` 参数（称其数值为 N）含义：N 大于 1 时，则每第 N 离子步的数据会被解析读取 
`ionic_step_offset` 参数含义：偏移量，与 `ionic_step_skip` 参数配合使用，表示从第几个离子步开始读取数据

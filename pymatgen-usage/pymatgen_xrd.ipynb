### 使用 pymatgen xrd 模块绘制 XRD 图谱

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.core.structure import Structure

structure = Structure.from_prototype("CsCl", ["Cs", "Cl"], a=4.2)
structure
```

```python
from spt.plot_params import set_roman_plot_params

calculator = XRDCalculator()

set_roman_plot_params()

calculator.get_plot(structure)
```

```python
calculator.get_pattern(structure).hkls
```

```python
calculator.get_pattern(structure).d_hkls
```

```python
from pymatgen.core.structure import Structure


structure_fn = "Nb5Si3_alpha.vasp"
structure = Structure.from_file(structure_fn)
structure
```

```python
structure.lattice
```

```python
lattice_constants = structure.lattice.abc
lattice_constants
```

```python
indices_nb = [i for i, site in enumerate(structure) if site.species_string == "Nb"]
indices_si = [i for i, site in enumerate(structure) if site.species_string == "Si"]
print(indices_nb)
print(indices_si)
```

```python
structure.get_distance(0, 1)
```

```python
r1 = lattice_constants[0] / 2
structure.get_neighbors(structure[10], r1)
```

```python
neighbors = structure.get_neighbors(structure[10], r1)
print(len(neighbors))

for neighbor in neighbors:
    index = neighbor.index
    distance = structure.get_distance(10, index)
    print(index, distance)
```

```python
r2 = lattice_constants[-1] / 2
neighbors = structure.get_neighbors(structure[10], r2)
print(len(neighbors))

for neighbor in neighbors:
    index = neighbor.index
    distance = structure.get_distance(10, index)
    print(index, distance)
```

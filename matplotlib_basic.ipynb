### matplotlib 科研绘图

#### 函数式绘图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label="$y=\sin(x)$")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

#### 对象式绘图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y, label="$y=\sin(x)$")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()

plt.show()
```

#### 3d 绘图

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10, 8))

ax.scatter(x, y, z, c="b", marker="o", label="3d plot")

ax.set(xlabel="X", ylabel="Y", zlabel="Z")
ax.legend()

plt.show()
```

#### 绘图示例

```python
"""
较美观的点线图

>[sci作图 - 计算材料学](https://ywwang0.github.io/2020/10/07/sci%E4%BD%9C%E5%9B%BE/)
"""

import matplotlib.pyplot as plt
from spt.plot_params import set_roman_plot_params

set_roman_plot_params(legend_labelspacing=0.4, legend_handletextpad=0.4)
fig, ax = plt.subplots()

x = [1, 2, 3, 4, 5]
y = [0.5456, 0.5919, 0.7318, 0.7362, 0.7203]

ax.plot(
    x,
    y,
    "^",
    c="g",
    ls="solid",
    lw=1,
    markeredgecolor="black",
    markersize=8,
    label="Test dataset",
)

ax.set_xlabel("Number of features")
ax.set_ylabel("$\mathregular{R^{2}}$")

ax.legend(loc="best", fancybox=False, edgecolor="black")
plt.show()
```

```python
# https://zhuanlan.zhihu.com/p/690957592

import matplotlib.pyplot as plt
import numpy as np
from spt.plot_params import set_roman_plot_params

from matplotlib.ticker import MultipleLocator

marker_list = [
    "o",
    "v",
    "^",
    "<",
    ">",
    "s",
    "p",
    "*",
    "h",
    "D",
]
color_list = [
    "k",
    "r",
    "coral",
    "orange",
    "palegreen",
    "deepskyblue",
    "lightsteelblue",
    "navy",
    "blueviolet",
    "pink",
]


label_list = [
    "Nature",
    "Science",
    "$Phys.Rev.Lett.$",
    "PNAS",
    "$J.Am.Chem.Soc.$",
    "Angew.Chem.Int.Ed",
    "H$^{+}$",
    "C$_{60}$",
    "Chelsea Football Club",
    "McLaren F1",
]

x = np.linspace(0, 1.0, 10)
y = []

set_roman_plot_params(legend_labelspacing=0.4, legend_handletextpad=0.4)
fig, ax = plt.subplots()

for i in range(0, 10):
    y.append(x**2 + 1 * i)
    ax.plot(
        x,
        y[i],
        c=color_list[i],
        marker=marker_list[i],
        mfc="None",
        ms=10,
        ls="--",
        lw=1,
        label=label_list[i],
    )

ax.legend(bbox_to_anchor=(1.0, 0.6), fontsize=11)

ax.xaxis.set_major_locator(MultipleLocator(0.2))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import pandas as pd
from spt.plot_params import set_roman_plot_params

color_list = [
    "#9900FF",
    "#4400FF",
    "#007DFF",
    "#00FF00",
    "#808000",
    "#FF6600",
    "#FF0000",
    "#000000",
]

xlsx_fn = "41563_2023_1702_MOESM5_ESM.xlsx"
df = pd.read_excel(xlsx_fn, sheet_name="Fig.3a", skiprows=2, header=None)

set_roman_plot_params()
fig, ax = plt.subplots()

for i in range(4):
    ax.plot(df.iloc[:, 2 * i], df.iloc[:, 2 * i + 1], color=color_list[i])

ax.set_xlim(7.2, 6.6)
ax.set_ylim(50000, 530000)
ax.set_yticks([])
ax.set_xlabel("Chemical shift (ppm)")

ax.xaxis.set_major_locator(MultipleLocator(0.2))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))

fontsize1 = 15
fontsize2 = 12
ax.text(7.18, 140000, "60C, electrolyte", color="#9900FF", fontsize=fontsize1)
ax.text(7.18, 80000, "60C, gas phase", color="#4400FF", fontsize=fontsize1)
ax.text(7.18, 270000, "200C, electrolyte", color="#007DFF", fontsize=fontsize1)
ax.text(7.18, 210000, "200C, gas phase", color="#00FF00", fontsize=fontsize1)
ax.text(7.0, 279000, "$^{14}$NH$_{4}$$^{+}$", fontsize=fontsize2)
ax.text(6.91, 279000, "$^{14}$NH$_{4}$$^{+}$", fontsize=fontsize2)
ax.text(6.82, 279000, "$^{14}$NH$_{4}$$^{+}$", fontsize=fontsize2)
ax.text(6.95, 500000, "$^{15}$NH$_{4}$$^{+}$", fontsize=fontsize2)
ax.text(6.86, 500000, "$^{15}$NH$_{4}$$^{+}$", fontsize=fontsize2)

plt.show()
```

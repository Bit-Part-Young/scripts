```python
# reference: https://github.com/K4ys4r/VASP_Scripts/blob/master/OSZICAR_MD_Plot/Plot_OSZICAR_MD.py

import numpy as np

file = "OSZICAR_MD"
with open(file, "r") as inp:
    f = inp.readlines()

data_list = []
for line in f:
   if "T=" in line:
       data = [float(x) for x in line.strip().split()[::2]]
       data_list.append(data)

data_array = np.array(data_list)

print(data_array.shape)
```

```python
import matplotlib.pyplot as plt
from spt.plot_params import set_roman_plot_params
from matplotlib.axes import Axes


set_roman_plot_params()
fig, ax = plt.subplots()
ax: Axes

ax.plot(data_array[:, 0], data_array[:, 1], label="Temperature")
```

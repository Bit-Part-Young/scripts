```python
import dpdata

poscar_fn = "Nb5Si3_alpha.vasp"
d_poscar = dpdata.System(poscar_fn, fmt="vasp/poscar")
d_poscar
```

```python
d_outcar = dpdata.LabeledSystem("OUTCAR", fmt="vasp/outcar")
d_outcar
```

```python
d_outcar["energies"]
```

```python
d_outcar["forces"]
```

```python
d_outcar["virials"]
```

转换成 lammps / vasp 格式，只能处理一帧？

```python
d_outcar.to("lammps/lmp", "outcar.lmp")
# d_outcar.to("lammps/lmp", "outcar.lmp", frame_idx=-1)
```

```python
dpdata.LabeledSystem("OUTCAR").to("lammps/lmp", "outcar2.lmp")
```

```python
dpdata.LabeledSystem("OUTCAR").sub_system([0, -1]).to("lammps/lmp", "outcar3.lmp")
```

```python
d_outcar.to("vasp/poscar", "CONTCAR")
```

转换成 deempd 格式，可处理全部帧的数据

```python
d_outcar.to("deepmd/raw", "dpmd_raw")
```
